```python
import pandas as pd
import requests
import json
from pprint import pprint
from sqlalchemy import create_engine
```

```python
#data from CSV file exported from FanGraphs
csv_file = "mets.csv"
mets_df = pd.read_csv(csv_file)
mets_df.head()
```

```python
#Cleaning up for the columns we want
nym_df = mets_df[['Name', 'HR', 'AVG', 'OBP', 'RBI', 'SB']].copy()
nym_df.head()
```

```python
url= "http://lookup-service-prod.mlb.com/json/named.search_player_all.bam?sport_code='mlb'&active_sw='Y'&start_season=2019&end_season=2019&name_part="
response = requests.get(url + "'"+nym_df['Name'][0]+"'")
```

```python
data = response.json()
pprint(data)
```

```python
nyminfo = []
i = 0
while i < len(nym_df['Name']):
    pInfoList = []
    pInfoList.append(nym_df['Name'][i])
    response = requests.get(url + "'"+nym_df['Name'][i]+"'")
    data = response.json()
    try:
        pInfoList.append(data['search_player_all']['queryResults']['row']['bats'])
        pInfoList.append(data['search_player_all']['queryResults']['row']['throws'])
        pInfoList.append(data['search_player_all']['queryResults']['row']['position'])
    except:
        pInfoList.append('NaN')
        pInfoList.append('NaN')
        pInfoList.append('NaN')
    nyminfo.append(pInfoList)
    i = i +1
    
```

```python
nymJson = pd.DataFrame(nyminfo)
nymJson.columns=['Name', 'Bats', 'Throws', 'Position']
nymJson.head()
```

```python
engine = create_engine(f'postgresql://postgres:postgres@127.0.0.1/baseball')


nym_df.to_sql(name='player_stats', con=engine, if_exists='append', index=False)
```

```python
nymJson.to_sql(name='player_info', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added by querying
pd.read_sql_query('select * from player_stats', con=engine).head()
```
